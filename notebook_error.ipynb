## 2. Error Analysis (Core Task)

```python
x_target = np.pi / 4
exact = np.cos(x_target)
h_values = np.logspace(-16, 0, 100)

errors_central = []
errors_richardson = []

for h in h_values:
    # Central Difference Error
    val_c = central_difference(np.sin, x_target, h)
    err_c = abs(val_c - exact) / abs(exact)
    errors_central.append(err_c)
    
    # Richardson Error
    val_r = richardson_extrapolation(np.sin, x_target, h)
    err_r = abs(val_r - exact) / abs(exact)
    errors_richardson.append(err_r)

errors_central = np.array(errors_central)
errors_richardson = np.array(errors_richardson)
```

```python
plt.figure(figsize=(10, 6))

# Plot Central Difference Error
plt.loglog(h_values, errors_central, 'b.-', label='Central Diff ($O(h^2)$)', alpha=0.6)

# Plot Richardson Error
plt.loglog(h_values, errors_richardson, 'r.-', label='Richardson ($O(h^4)$)', alpha=0.6)

plt.xlabel('Step Size $h$')
plt.ylabel('Relative Error')
plt.title('Comparison of Numerical Differentiation Methods')
plt.legend()
plt.grid(True, which="both", alpha=0.3)
plt.show()
```
